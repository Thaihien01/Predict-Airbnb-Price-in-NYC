```python
def outliers(data, feature):
    q1,q3 = data[feature].quantile(0.25), data[feature].quantile(0.75)
    iqr = q3-q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    rows=data.loc[(data[feature] < lower) | (data[feature] > upper), feature]=data[feature].median()    
    print(rows)
```

```python
def remove_outliers(data, feature):
    title = outliers(data, feature)
    sns.boxplot(data[feature])
    plt.title(title)
    plt.show()
```

```python
import sklearn
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.linear_model import Lasso
```

```python
en = preprocessing.LabelEncoder()  
en.fit(df['neighbourhood_group'])
df['neighbourhood_group'] = en.transform(df['neighbourhood_group'])

en = preprocessing.LabelEncoder()
en.fit(df['neighbourhood'])
df['neighbourhood'] = en.transform(df['neighbourhood'])

en = preprocessing.LabelEncoder()
en.fit(df['room_type'])
df['room_type'] = en.transform(df['room_type'])

df.sort_values(by='price', ascending=True, inplace=True)

df.head()
```

```python
X=df.drop(['price'], axis=1)
y=df['price']
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=34)
X_train.shape
```

# LinearRegression

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

# Decision Tree

```python
model = DecisionTreeRegressor(min_samples_leaf=.001)
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test,y_test))
```

# Lasso Regression

```python
model = Lasso(alpha = .001)
model.fit(X_train, y_train)
print(model.score(X_train,y_train))
print(model.score(X_test, y_test))
```

# Random Forest

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
print(model.score(X_train, y_test))
print(model.score(X_test, y_test))
```
